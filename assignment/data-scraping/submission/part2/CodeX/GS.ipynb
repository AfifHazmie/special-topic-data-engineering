```python
import requests
from bs4 import BeautifulSoup

search_terms = ["school of computing universiti teknologi malaysia", "faculty of computing universiti teknologi malaysia"]

for term in search_terms:
    print(f"Results for '{term}':\n")
    # Perform a search query and get the top 50 results
    url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=%22Universiti+Teknologi+Malaysia%22+author%3AUTM&btnG="
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all('div', {'class': 'gs_ri'})
    for i in range(50):
        try:
            result = results[i]
            title = result.find('h3', {'class': 'gs_rt'}).text.strip()
            authors = result.find('div', {'class': 'gs_a'}).text.strip()
            link = result.find('a')['href']
            year = result.find('div', {'class': 'gs_a'}).text.strip().split(',')[-1].strip()
            print(f"Title: {title}")
            print(f"Authors: {authors}")
            print(f"Year: {year}")
            print(f"Link: {link}\n")
        except IndexError:
            break
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://scholar.google.com/scholar?hl=en&q=%22school+of+computing+universiti+teknologi+malaysia%22+OR+%22faculty+of+computing+universiti+teknologi+malaysia%22&btnG=&as_sdt=1%2C5&as_sdtp="

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

for i in soup.select('.gs_r'):
    title = i.select_one('.gs_rt a')
    if title:
        print("Title:", title.text.strip())
    else:
        print("Title: N/A")

    author = i.select_one('.gs_a')
    if author:
        print("Author(s):", author.text.strip())
    else:
        print("Author(s): N/A")

    link = i.select_one('.gs_rt a')['href']
    if link:
        print("Link:", link)
    else:
        print("Link: N/A")

    cited_by = i.select_one('.gs_fl a:nth-of-type(1)')
    if cited_by:
        print("Cited by:", cited_by.text.split(' ')[-1])
    else:
        print("Cited by: N/A")

    year = i.select_one('.gs_a')
    if year:
        year = year.text.strip().split(',')[-1].strip()
        print("Year:", year)
    else:
        print("Year: N/A")

    print("="*50)

```

```python
import json
import requests
import pandas as pd

api_key = "592b22c1ce7080ab7e48819dd55666563ffa906acd09c3eb0f36d8683bc2ffe1"
params = {
    "engine": "google_scholar",
    "q": "school of computing universiti teknologi malaysia OR faculty of computing universiti teknologi malaysia",
    "api_key": api_key,
    "num": 20, # number of results to retrieve, default is 10
    "start": 0, # starting index of the results, default is 0
    "tbm": "nws", # search for news articles, default is "nws"
    "sort_by": "date", # sort by date, default is "relevance"
}

response = requests.get("https://serpapi.com/search.json", params=params)
data = json.loads(response.text)

results = []
for item in data["organic_results"]:
    result = {}
    result["title"] = item["title"]
    result["link"] = item["link"]
    result["snippet"] = item["snippet"]
    results.append(result)

df = pd.DataFrame(results)
print(df.head())

# Save to CSV
df.to_csv("results.csv", index=False)
```
